```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('23.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect corners using Harris Corner Detection
corners = cv2.cornerHarris(gray_image, 2, 3, 0.04)

# Threshold to obtain the corners
threshold = 0.01 * corners.max()
corner_image = image.copy()
corner_image[corners > threshold] = [0, 0, 255]  # Red color for corners

# Display the image with corners
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(corner_image, cv2.COLOR_BGR2RGB))
plt.title('Harris Corner Detection')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the two images
image1 = cv2.imread('23.jpg')
image2 = cv2.imread('23.jpg')


# Convert images to grayscale
gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Perform Harris corner detection on both images
harris_corners1 = cv2.cornerHarris(gray_image1, 2, 3, 0.04)
harris_corners2 = cv2.cornerHarris(gray_image2, 2, 3, 0.04)

# Threshold for an optimal value, it may vary depending on the image.
threshold = 0.01 * harris_corners1.max()

# Find keypoints in the images
keypoints1 = np.argwhere(harris_corners1 > threshold).astype(np.float32)
keypoints2 = np.argwhere(harris_corners2 > threshold).astype(np.float32)

# Initialize a Brute Force Matcher
bf = cv2.BFMatcher()

# Compute descriptors for keypoints
orb = cv2.ORB_create()
_, descriptors1 = orb.compute(gray_image1, [cv2.KeyPoint(x[1], x[0], 3) for x in keypoints1])
_, descriptors2 = orb.compute(gray_image2, [cv2.KeyPoint(x[1], x[0], 3) for x in keypoints2])

# Match descriptors of the two images
matches = bf.knnMatch(descriptors1, descriptors2, k=2)

# Apply ratio test to obtain good matches
good_matches = []
for m, n in matches:
    if m.distance < 0.75 * n.distance:
        good_matches.append(m)

# Draw matches
matched_image = cv2.drawMatches(image1, [cv2.KeyPoint(x[1], x[0], 3) for x in keypoints1], 
                                 image2, [cv2.KeyPoint(x[1], x[0], 3) for x in keypoints2], 
                                 [cv2.DMatch(i, i, 0) for i in range(len(good_matches))], None, 
                                 flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# Convert the image to RGB (OpenCV uses BGR by default)
matched_image_rgb = cv2.cvtColor(matched_image, cv2.COLOR_BGR2RGB)

# Display the matched image
plt.figure(figsize=(12, 8))
plt.imshow(matched_image_rgb)
plt.title(f'Number of Good Matches: {len(good_matches)}')
plt.axis('off')
plt.show()

```
